## Objective
Load all raw mutual fund datasets and verify their structure.

```python
import pandas as pd
import os

DATA_DIR = "../data/raw"

files = os.listdir(DATA_DIR)
print(files)
```

```python
for file in files:
    df = pd.read_csv(os.path.join(DATA_DIR, file))
    print(f"\n{file}")
    print(df.head())
    print(df.shape)
```
