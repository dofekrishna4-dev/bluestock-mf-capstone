# Data Cleaning

## Objective
Clean missing values, duplicates and standardize datasets before loading into SQLite.

```python
import pandas as pd

df = pd.read_csv("../data/raw/07_scheme_performance.csv")

df.head()
```

```python
print(df.isnull().sum())
```

```python
df = df.drop_duplicates()

df.columns = df.columns.str.lower().str.strip()

df.head()
```

```python
df.to_csv("../data/processed/07_scheme_performance.csv", index=False)
```
